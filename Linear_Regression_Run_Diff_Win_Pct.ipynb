```python
#Linear Regression Runs & Wins Predict Wins using Run Differential

teams = pd.read_csv('/content/Teams.csv')
```

```python
teams
```

```python
teams.shape
```

```python
#look at subset of teams data & filter  for 21st century teams only

tfcteams = teams[['yearID','teamID','lgID','G','W','L','R','RA']] [teams['yearID']>2000]
```

```python
#make two new variables: 1.Run Differential 2.Win%

tfcteams['RunDiff']=tfcteams['R'] - tfcteams['RA']
tfcteams['Winpct']= tfcteams['W'] / (tfcteams['W'] + tfcteams['L'])
```

```python
tfcteams
```

```python
#See relationship between  Run Differential & Win%
#also add Ordinary Least Squares(OLS) trendline

#Check model assunption: Linearity

plot = px.scatter(tfcteams, x='RunDiff', y='Winpct', trendline='ols')
plot.show()
```

```python
#train model to predict Win% & number of wins using run differential and Linear regression model: I am using statsmodel package 
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

statsmodels- a Python library for statistical modeling and inference, providing a wide range of tools for regression analysis, time series analysis, and hypothesis testing.

```python
#making model
model = smf.ols(formula = 'Winpct ~ RunDiff',data=tfcteams).fit()
```

```python
#Check Model Accuracy & fit using summary() function
#OLS Regression Results
#High R-Squared Value means strong relationship between Run Differential & Win%
#Low P value
model.summary()
```

# Interpretation of OLS Regression Results
The p-value for the coefficient estimate of the x variable (RunDiff) is 0.000, which is less than the typical significance level of 0.05. This suggests that the x variable is a significant predictor of the y variable (Winpct), meaning that there is a statistically significant linear relationship between a team's run differential and their win percentage.

The 95% confidence interval for the coefficient estimate of the x variable (RunDiff) is [0.001, 0.001], indicating that we can be 95% confident that the true population coefficient falls within this range. This means that, on average, a one-unit increase in run differential is associated with an increase in win percentage between 0.001 and 0.001.

```python
#Now predict WIn% and Wins using model I fit

tfcteams['predictedWinpct']=model.predict()
tfcteams['predictedW']= round(tfcteams['W']+tfcteams['L'])*tfcteams['predictedWinpct']
```

```python
tfcteams[['Winpct','predictedWinpct','W','predictedW']]
```

```python
#check normality of model by looking at residuals
#studentized residuals tend to be better at outlier testing

influence= model.get_influence()
residuals = influence.resid_studentized_internal
```

```python
#create graph of our residuals and highlight any outliers or values higher than 2.8
#Model assumption: Homoscedasticity
import numpy as np
import plotly.graph_objs as go

residplot = go.Figure(go.Scatter(
    x= tfcteams['RunDiff'],
    y=residuals,
    mode='markers',
    marker=dict(
        color=
        pd.Series(np.where(residuals>2.8,'outlier',np.where(residuals<-2.8,'outlier','normal')))
        .astype(str).map({'normal':0, 'outlier':1}),
        colorscale =[[0,'blue'],[1,'red']]
    )
)

)

residplot.show()
```
